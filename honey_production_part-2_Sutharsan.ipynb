# 3. Display the percentage distribution of the data in each year using the pie chart.

```python
# Group the data by year and calculate the sum of totalprod for each year
yearly_totalprod = hp.groupby('year')['totalprod'].sum()
```

```python
yearly_totalprod
```

```python
# Create a pie chart
plt.figure(figsize=(8,6))
plt.pie(yearly_totalprod, labels=yearly_totalprod.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage Distribution of Total Honey Production by Year')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Show the pie chart
plt.show()
```

# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Create a displot (distribution plot)
plt.figure(figsize=(8, 6))
sns.displot(hp['priceperlb'], kde=True);

# Add title and labels
plt.title('Distribution of Price per Pound of Honey')
plt.xlabel('Price per Pound')
plt.ylabel('Frequency');
```

The distribution of the variable "price per lb" using displot findings:

* The price per pound of honey is normally distributed.

* Mean 1.4095686900958466 and  Standard Deviation 0.6385989458761435.

* The price per pound of honey is slightly skewed to the right.

* which means that there are more prices at the lower end of the distribution than at the higher end.

* The distribution is relatively smooth, which means that there are no large gaps in the data.

* The distribution is not perfectly normal, but it is close enough to be useful for understanding the price of honey.

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
# Create a scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(hp['numcol'], hp['prodvalue'], alpha=0.7)

# Add title and labels
plt.title('Relationship between Number of Colonies and Production Value')
plt.xlabel('Number of Colonies')
plt.ylabel('Production Value')

# Show the plot
plt.show()
```

The relationship between the variables define our findings:
* There is a positive correlation between the number of colonies and the production value. which means that as the number of colonies increases, the production value also increases.

* The scatterplot shows that there is a positive correlation between the number of colonies and the production value. This means that more colonies can lead to higher production values.

* The overall scatterplot is positive correlation between the colonies and production value to spread of this  data. 

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings. 

```python
# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='year', y='prodvalue', data=hp)

# Add title and labels
plt.title('Relationship between Year and Production Value (Boxplot)')
plt.xlabel('Year')
plt.ylabel('Production Value');
```

The relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot findings:
* The interquartile range (IQR) has been increasing over the past few years. This means that the spread of the data is increasing.
* The number of outliers has been decreasing over the past few years. This means that the data is becoming more evenly distributed.


# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
# Select relevant columns for pairplot
columns_for_pairplot = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year']

# Create a pairplot
sns.pairplot(hp[columns_for_pairplot], hue='year', diag_kind='kde', markers='o')

# Add title to the pairplot
plt.suptitle('Pairplot of Selected Variables by Year', y=1.02)

# Show the pairplot
plt.show()
```

#Inferences:
* There is a positive correlation between the numcol and totalprod the production value.

* There is a positive correlation between the yield per colony and the production value.

* There is a negative correlation between the total production and the production value.

* The relationship between the variables various by year.

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
# Select relevant columns for correlation analysis
columns_for_correlation = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']

# Calculate the correlation matrix
correlation_matrix = hp[columns_for_correlation].corr()
```

```python
correlation_matrix
```

```python
# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', center=0)

# Add title to the heatmap
plt.title('Correlation Heatmap')

# Show the heatmap
plt.show()
```

The correlation values using a plot in inference:
* There is a strong positive correlation between the number of colonies and the total production to the production value.

* There is a weak negative correlation between the stock variables and the price per ib.

* There is a weak positive correlation between the price per pound and the production value.
